#### Parameter Tuning

```python
import pandas as pd
import numpy as np

dataset = pd.read_csv('Churn_Modelling.csv')
X = dataset.iloc[:, 3:13].values
y = dataset.iloc[:, 13].values


from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer


# Male/Female
le_X = LabelEncoder()
X[:, 2] = le_X.fit_transform(X[:, 2])

# Country column
ct = ColumnTransformer([("Country", OneHotEncoder(), [1])], remainder = 'passthrough')
X = ct.fit_transform(X)


X = X[:,1:]


# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)

# standardizing all the variable into similar scale of values
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)

# Fitting classifier to the Training set
# Create your classifier here
import keras

#sequential model to initialize the neural networks and dense module to build the layers for NN
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from sklearn.metrics import confusion_matrix
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score,GridSearchCV

def build_classifier(optimizer,activation):
    cls = Sequential()
    cls.add(Dense(output_dim = 6,init = 'uniform',  activation = activation, input_dim = 11))
    #cls.add(Dropout(p=0.1))
    cls.add(Dense(output_dim = 6, init = 'uniform', activation = activation))
    cls.add(Dense(output_dim = 6, init = 'uniform', activation = activation))
    #cls.add(Dropout(p=0.1))
    cls.add(Dense(output_dim = 1, init = 'uniform', activation = 'sigmoid'))
    cls.compile(optimizer = optimizer,loss = 'binary_crossentropy',metrics = ['accuracy'])
    return(cls)

cls = KerasClassifier(build_fn = build_classifier)

gridParameters = {'batch_size':[18,20,22,25],
                  'epochs' : [400,500],
                  'optimizer' : ['adam','rmsprop'],
                  'activation': ['selu','relu','elu']
                 }

gridSearch = GridSearchCV(cls,param_grid=gridParameters,scoring = 'accuracy',cv = 10,n_jobs = -1)
gridSearch.fit(X_train,y_train)

bestParam = gridSearch.best_params_
bestAccuracy = gridSearch.best_score_

print("best params and accuracy from the gridsearchcv are {}, {} respectively".format(bestParam,bestAccuracy))
```
